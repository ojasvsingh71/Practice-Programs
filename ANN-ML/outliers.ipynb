# Outliers
Data points that deviate significantly from the general distribution of the dataset.

- Impact on ML models:  Can bias parameter estimation
- Sources: Data errors 
- Handling method: Removal, transformation or robust algorithms less sensitive to outliers.
- Domain dependance: What is considered an oulier in one domain may be normal in another.

## Categories:

1. Global Outliers:  
- Individual data points that lie far from the rest.
2. Contextual Outliers: 
- Ouliers relative to specific context or condition.
3. Collective Outliers: 
- A group of related data points behaving anomalously together.
