```python
import numpy as np

# Q1
visitors = np.array([20, 21, np.nan, 22, 210, 19, 20])

# (a) Replace missing value with median
median_val = np.nanmedian(visitors)
visitors_filled = np.where(np.isnan(visitors), median_val, visitors)
print("Median:", median_val)
print("After replacing NaN:", visitors_filled)

# (b) Detect outlier using IQR
Q1 = np.percentile(visitors_filled, 25)
Q3 = np.percentile(visitors_filled, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

outliers = visitors_filled[(visitors_filled < lower_bound) | (visitors_filled > upper_bound)]
print("Outliers:", outliers)
```

```python
# Q2
transactions = np.array([10, 11, 12, 13, 150])
mean_val = np.mean(transactions)
std_val = np.std(transactions)
z_score_150 = (150 - mean_val) / std_val

print("\nMean:", mean_val)
print("Std Dev:", std_val)
print("Z-score of 150:", z_score_150)
```
