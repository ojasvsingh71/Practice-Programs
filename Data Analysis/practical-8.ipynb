```python
from collections import defaultdict

def mapper_A(A, rows_A, cols_A, cols_B):
    mapped = []

    for i in range(rows_A):
        for j in range(cols_A):
            for k in range(cols_B):

                # Key = (i, k)
                # Value = ("A", j, A[i][j])
                mapped.append(((i, k), ("A", j, A[i][j])))

    return mapped


def mapper_B(B, rows_A, cols_B, rows_B):
    mapped = []

    for j in range(rows_B):
        for k in range(cols_B):
            for i in range(rows_A):

                # Key = (i, k)
                # Value = ("B", j, B[j][k])
                mapped.append(((i, k), ("B", j, B[j][k])))

    return mapped
```

```python
def reducer(grouped_data):
    result = {}

    for key, values in grouped_data.items():

        A_values = {}
        B_values = {}

        # Separate A and B values
        for value in values:

            matrix_name, j, val = value

            if matrix_name == "A":
                A_values[j] = val

            else:
                B_values[j] = val

        # Compute multiplication sum
        total = 0

        for j in A_values:
            if j in B_values:
                total += A_values[j] * B_values[j]

        result[key] = total

    return result
```

```python
A = [
    [1, 2],
    [3, 4]
]

B = [
    [5, 6],
    [7, 8]
]

rows_A = len(A)
cols_A = len(A[0])

rows_B = len(B)
cols_B = len(B[0])

# Step 1: Map Phase
mapped_A = mapper_A(A, rows_A, cols_A, cols_B)
mapped_B = mapper_B(B, rows_A, cols_B, rows_B)

mapped_data = mapped_A + mapped_B

# Step 2: Shuffle and Sort Phase
grouped_data = defaultdict(list)

for key, value in mapped_data:
    grouped_data[key].append(value)

# Step 3: Reduce Phase
result = reducer(grouped_data)

# Step 4: Display Result Matrix
C = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

for (i, k), value in result.items():
    C[i][k] = value

print("Result Matrix:")

for row in C:
    print(row)
```
