# Naive Bayes


## Conditional Probability

P(A and B) = P(A) * P(B/A)

P(B and A) = P(B) * P(A/B)

P(A) * P(B/A) = P(B) * P(A/B)  

P(B/A) = (P(B) * P(A/B))/P(A) → Bayes' Theorem

P(y/x1,x2,x3...) = (P(y) * P(x1,x2,x3.../y))/P(x1,x2,x3...)
