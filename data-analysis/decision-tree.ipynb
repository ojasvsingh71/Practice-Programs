```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

age = ["Young","Young","Middle","Old","Old","Old","Middle","Young"]
income = ["High","High","High","Medium","Low","Low","Low","Medium"]
student = ["No","No","No","No","Yes","Yes","Yes","No"]
buy = ["No","No","Yes","Yes","Yes","No","Yes","No"]
```

```python
le = LabelEncoder()

age_enc = le.fit_transform(age)
income_enc = le.fit_transform(income)
student_enc = le.fit_transform(student)
buy_enc = le.fit_transform(buy)
```

```python
X = list(zip(age_enc, income_enc, student_enc))
y = buy_enc

model = DecisionTreeClassifier(criterion="entropy")
model.fit(X, y)
```

```python
print(model.predict([[0, 0, 0]])) 
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plot_tree(model, feature_names=["Age","Income","Student"], class_names=["No","Yes"], filled=True)
plt.show()
```
